```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
tweet=pd.read_csv('/content/twitter_training.csv')
tweet.head(10)
```

```python
column=['ID','Entity','Sentiment','Content']
tweet=pd.read_csv('/content/twitter_training.csv',names=column)
tweet
```

```python
tweet.shape
```

```python
tweet.isnull().sum().sum()
```

```python
tweet.dropna(axis=0,inplace=True)
```

```python
tweet.isnull().sum().sum()
```

```python
tweet.duplicated().sum()
```

```python
tweet.drop_duplicates(inplace=True)
tweet.duplicated().sum()
```

```python
tweet.shape
```

```python
sentiment_count=tweet['Sentiment'].value_counts()
sentiment_count
```

```python
plt.figure(figsize=(6,3))
sentiment_count.plot(kind='bar',color=['blue','green','red','yellow'])
plt.title('Sentiment Count')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
brand_data=tweet[tweet['Entity'].str.contains('Microsoft',case=False)]
brand_sentiment_counts=brand_data['Sentiment'].value_counts()
brand_sentiment_counts
```

```python
plt.figure(figsize=(6,6))
plt.pie(brand_sentiment_counts,labels=brand_sentiment_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Brand Sentiment Analysis')
plt.show()
```
